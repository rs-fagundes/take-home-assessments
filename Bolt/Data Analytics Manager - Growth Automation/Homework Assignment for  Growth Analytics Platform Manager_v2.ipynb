```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lood dataframes
df_supply = pd.read_csv('csv/Supply Data.csv')
df_demand = pd.read_csv('csv/Demand Data.csv')


# merge dataframes
df_supply.rename(columns={'Date': 'Date_Hour'}, inplace=True)
df_demand.rename(columns={'Date': 'Date_Hour'}, inplace=True)
df_supply_demand = pd.merge(df_supply, df_demand, on='Date_Hour', how='left')


# Split Date_Hour field into Date and Hour
df_supply_demand[['Date', 'Hour']] = df_supply_demand['Date_Hour'].str.split(" ", expand=True)
df_supply_demand['Date'] = pd.to_datetime(df_supply_demand['Date'])
df_supply_demand['Hour'] = df_supply_demand['Hour'].astype(int)


# Extract Weekday and Month from Date
df_supply_demand['Weekday'] = df_supply_demand['Date'].dt.weekday  # (Field ID 4) Day of the week (0 = Monday)
df_supply_demand['Weekday Name'] = df_supply_demand['Date'].dt.day_name()  # (Field ID 5) Day name (e.g., Monday)
df_supply_demand['Month'] = df_supply_demand['Date'].dt.month  # (Field ID 6) Numeric month
df_supply_demand['Month Name'] = df_supply_demand['Date'].dt.month_name()  # (Field ID 7) Month name (e.g., December)


# (Field ID 18) Total passenger demand
df_supply_demand['Demand'] = df_supply_demand['People saw 0 cars (unique)'] + df_supply_demand['People saw +1 cars (unique)']


# (Field ID 19) Total supply (drivers visible to passengers)
df_supply_demand['Supply'] = df_supply_demand['People saw +1 cars (unique)']


# (Field ID 20) Coverage per Online (Coverage per online hour)
df_supply_demand['Coverage per Online'] = df_supply_demand['Coverage Ratio (%)'] / df_supply_demand['Online (h)']


# (Field ID 21) Coverage Deficit (Deficit in coverage)
coverage_threshold = 100  # Coverage target
df_supply_demand['Coverage Deficit'] = coverage_threshold - df_supply_demand['Coverage Ratio (%)']


# (Field ID 22) Additional Hours Needed (Hours needed to reach full coverage)
df_supply_demand['Additional Hours Needed'] = df_supply_demand['Coverage Deficit'] / df_supply_demand['Coverage per Online']


# (Field ID 23) New Coverage Ratio (Expected coverage with additional hours)
df_supply_demand['New Coverage Ratio'] = (
    df_supply_demand['Coverage per Online'] *
    (df_supply_demand['Online (h)'] + df_supply_demand['Additional Hours Needed'])
)

# (Field ID 24) Additional Rides (Additional rides expected from missed coverage)
df_supply_demand['Additional Rides'] = df_supply_demand['People saw 0 cars (unique)']


# (Field ID 25) Total Revenue (€) (Total revenue generated from rides)
finished_ride_value = 10  # Average value of a finished ride
df_supply_demand['Total Revenue'] = df_supply_demand['Finished Rides'] * finished_ride_value


# (Field ID 26) Driver Income (€) (80% of total revenue earned by drivers)
driver_share = 0.8  # Share percentage for drivers
df_supply_demand['Driver Income'] = df_supply_demand['Total Revenue'] * driver_share


# (Field ID 27) Hour Rate (€) (Income per online hour for drivers)
df_supply_demand['Hour Rate'] = df_supply_demand['Driver Income'] / df_supply_demand['Online (h)']


# (Field ID 28) Guaranteed Income (€) (Guaranteed income provided to drivers)
df_supply_demand['Guaranteed Income'] = (
    df_supply_demand['Hour Rate'] * 
    (df_supply_demand['Online (h)'] + df_supply_demand['Additional Hours Needed'])
)


# (Field ID 29) Compensation (€) (Compensation paid by the company if income is below guaranteed income)
df_supply_demand['Compensation'] = df_supply_demand['Guaranteed Income'] - df_supply_demand['Driver Income']
df_supply_demand['Compensation'] = df_supply_demand['Compensation'].clip(lower=0)
```

```python
# Fix any missing or null values
df_supply_demand.fillna(0, inplace=True)
df_supply_demand.replace([float('inf'), float('-inf')], 0, inplace=True) 
```

<h2>1. Critical Undersupply Periods: Analysis and Insights for Optimized Driver Availability</h2>

```python
# Calculate average Supply and Demand
avg_demand_supply = df_supply_demand.groupby('Hour').agg({
    'Supply': 'mean',
    'Demand': 'mean'
}).reset_index()

# Calculate difference betwen Demand and Supply
avg_demand_supply['Gap'] = avg_demand_supply['Demand'] - avg_demand_supply['Supply']

# Calcultae distance variable to be used on lineplot
distance = np.abs(avg_demand_supply['Gap'])

# Define a peakhour threshold (above 60%)
threshold = avg_demand_supply['Gap'].quantile(0.60)

sns.set_theme(style="white", context="notebook")

# Create lineplot
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=avg_demand_supply, x='Hour', y='Demand', 
    label='Average Demand', marker='s', color="blue"
)
sns.lineplot(
    data=avg_demand_supply, x='Hour', y='Supply', 
    label='Average Supply', marker='o', color="green"
)
plt.fill_between(
    avg_demand_supply['Hour'], 
    avg_demand_supply['Demand'], 
    avg_demand_supply['Supply'], 
    where=(distance > threshold),
    color="red", alpha=0.3, label="Distance (Gap)", hatch="|"
)

plt.title('24-hour curve of average demand and supply volume', fontsize=14, loc="left", pad=10)
plt.xlabel('Hours', fontsize=12)
plt.ylabel('Average Volume', fontsize=12)
plt.xticks(range(0, 24, 1), fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10, frameon=False)
sns.despine() 
plt.tight_layout()
plt.show()
```

```python
# Calculate average Coverage Ratio
coverage_by_weekday_hour = df_supply_demand.groupby(['Weekday', 'Weekday Name', 'Hour'])['Coverage Ratio (%)'].mean().round(0).unstack()


plt.figure(figsize=(10, 5))
sns.heatmap(
    coverage_by_weekday_hour, 
    cmap="Greens", 
    annot=True, 
    fmt=".0f", 
    cbar=True,
    annot_kws={"size": 8},
    linewidths=0.5
)

plt.title('Weekly Coverage Ratio per Hour (%)', fontsize=14, loc='left')
plt.xlabel('Hours', fontsize=10)
plt.ylabel('Weekday', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9, rotation=0)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

<h2>2. Optimizing Coverage: Required Online Hours for Peak Periods</h2>

```python
# Calculate Average Additional Hours Needed
avg_additional_hours_by_weekday_hour = df_supply_demand.groupby(['Weekday', 'Weekday Name', 'Hour'])['Additional Hours Needed'].mean().round(0).unstack()

plt.figure(figsize=(10, 5))
sns.heatmap(
    avg_additional_hours_by_weekday_hour, 
    cmap="Blues", 
    annot=True, 
    fmt=".0f", 
    cbar=True,
    annot_kws={"size": 8},
    linewidths=0.5
)

plt.title('Additional Hours Needed', fontsize=14, loc='left')
plt.xlabel('Hours', fontsize=10)
plt.ylabel('Weekday', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9, rotation=0)
sns.despine(left=True, bottom=True) 
plt.tight_layout()
plt.show()
```

<h2>3. Guaranteed Earnings Analysis: Attracting Drivers During Peak Hours</h2>

```python
# Peak Hours Array
peak_hours = [3,4, 7, 8, 9, 17, 18, 19, 20]

# Filter dataframe with peak hours only
df_peak_hours = df_supply_demand[df_supply_demand['Hour'].isin(peak_hours)]

# Calculate average compensation
avg_compensation_by_weekday_hour = df_peak_hours.groupby(['Weekday', 'Weekday Name', 'Hour'])['Compensation'].mean().round(0).unstack()

plt.figure(figsize=(10, 5))
sns.heatmap(
    avg_compensation_by_weekday_hour, 
    cmap="Purples", 
    annot=True, 
    fmt=".0f", 
    cbar=True,
    annot_kws={"size": 8},
    linewidths=0.5
)

plt.title('Compensation per Hour (Peak Hours)', fontsize=14, loc='left')
plt.xlabel('Peak Hours', fontsize=10)
plt.ylabel('Weekday', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9, rotation=0)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
